**ADVANCED NUMPY**

```python
#numpy arrays vs python lists
#python - takes more time in execution
#dynamic array
#referential array - address of the item is stored thats why time is taken more
a=[i for i in range(10000000)]
b=[i for i in range(10000000,20000000)]

c=[]
import time

start=time.time()
for i in range(len(a)):
  c.append(a[i]+b[i])

print(time.time()-start)
```

```python
#numpy - takes less time in execution - faster than python sequences
import numpy as np
a=np.arange(10000000)
b=np.arange(10000000,20000000)

start=time.time()
c=a+b
print(time.time()-start)
```

```python
#memory
a =[i for i in range(10000000)]
import sys #system

sys.getsizeof(a)
```

```python
#memory-numpy
a=np.arange(10000000)
sys.getsizeof(a)
```

Numpy:
takes less time ,
takes less memory ,
more convenient to work with

**BROADCASTING**

describes how numpy array treat two arrays with different shapes while performing arithmetic operations....

The smaller array is "broadcasted" across the larger array so that they have compatible shapes

```python
#same shape
a=np.arange(6).reshape(2,3)
b=np.arange(6,12).reshape(2,3)
print(a)
print(b)

print(a+b) #item wise addition
```

```python
#diff shape
a=np.arange(6).reshape(2,3)
b=np.arange(3).reshape(1,3)
print(a)
print(b)

print(a+b) #
```

```python
#examples
a=np.arange(12).reshape(4,3)
b=np.arange(3)

print(a)
print(b)

print(a+b)
```

```python
#ex
a=np.arange(3).reshape(1,3)
a=np.arange(3).reshape(3,1)

print(a)
print(b)

print(a+b)
```

```python
a=np.arange(3).reshape(1,3)
b=np.arange(4).reshape(4,1)

print(a)
print(b)

print(a+b)
```

```python
a=np.array([1]) #shape = (1,1)
b=np.arange(4).reshape(2,2) #shape=(2,2)

print(a)
print(b)

print(a+b)
```

```python
#sigmoid : used in Logistic regression -ML
def sigmoid(array):
  return 1/(1+np.exp(-(array)))

a = np.arange(10)

sigmoid(a)
```

```python
#mean squared error

actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)

def mse(actual,predicted):
  return np.mean((actual-predicted)**2)

mse(actual,predicted)


```
